## <a id="data_prep"/> Data Preparation

In this section, I am going to separate and clean the data in different subdatasets, in order to answer the business questions.  

```python
columns = df.columns
print(columns)
```

```python
#The first question: Where do most accidents happen
df_where = df[['ID','Start_Lat', 'Start_Lng', 'End_Lat', 'End_Lng','City', 'County', 'State',
       'Zipcode', 'Country', 'Timezone', 'Airport_Code', 'Severity']]
```

In this first question, we want to consider every available information, so we are not going to remove any missing data from this columns

```python
#The second question: The surroundings of the accidents
df_surr = df[['Amenity', 'Bump', 'Crossing',
       'Give_Way', 'Junction', 'No_Exit', 'Railway', 'Roundabout', 'Station',
       'Stop', 'Traffic_Calming', 'Traffic_Signal', 'Turning_Loop']]
```

From the infomation above, we know that these columns do not have missing data

```python
#The third question: When do most accidents happen?
df_when = df[['ID', 'Start_Time', 'End_Time', 'Severity']]
```

For this question, we also know there is no missing data. Nonetheless, we have to separate the date and time of the accident, to answer the question.

```python
df_when['Start_Date'] = pd.to_datetime(df_when['Start_Time']).dt.date
df_when['Start_Time_1'] = pd.to_datetime(df_when['Start_Time']).dt.time
df_when['End_Date'] = pd.to_datetime(df_when['End_Time']).dt.date
df_when['End_Time_1'] = pd.to_datetime(df_when['End_Time']).dt.time

```

```python
print(df_when)
```

```python
#Create hour column to group the data by hour of the day 
df_when['hour'] = pd.to_datetime(df_when['Start_Time']).dt.strftime('%H')
```

```python
df_when['hour']
```

```python
#Create month column to group the data by month 
df_when['month'] = pd.to_datetime(df_when['Start_Date']).dt.strftime('%b')
```

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", 
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
#Turn the months into numerical values to present them in order
df_when['month'] = pd.Categorical(df_when['month'], categories=months, ordered=True)
df_when['month']
```

```python
#Create month column to group the data by month 
df_when['year'] = pd.to_datetime(df_when['Start_Date']).dt.strftime('%Y')
df_when['year']
```

```python
#The final question: What relates to accident severity?

#Extract to a new df the numerical variables
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']

df_severity = df.select_dtypes(include=numerics)
```

Here, we are going to focus on numerical values, to see if they have some correlation to severity

```python
#Remove the columns in question
df_severity = df_severity.drop(columns = ['Start_Lat', 'Start_Lng', 'End_Lat','End_Lng','Number'])
```

From those numerical values, we're going to remove the coordinates and the number, as the dataset of the 'where' question, can also be used to map severity

```python
columns = df_severity.columns
print(columns)
for column in columns:
    print(column)
    print(df_severity[column].isna().mean())
    print()
```

From these number, we can see there are many missing values. We are going to remove them, in order to see how the accidents that have those columns filled correlate to the accident severity

```python
#Removing missing data from those columns
df_severity = df_severity.dropna()
```

```python
print(columns)
for column in columns:
    print(column)
    print(df_severity[column].isna().mean())
    print()
```

## <a id="description"/> What are the most common words in an accident description?

```python
def clean_and_plot_words(description):
    """Removes Nan from the description column. Takes the descriptions from the accident description column 
     and joins them in a list. Converts the list into a string. Finally, returns a plot with the top 500 used words in the description in 
     a Word Cloud plot 
    
    Parameters
    ----------
    description : series
        The dataframe column with the accidents description.

    Returns
    -------
    A Word Cloud plot with the 500 most used words in the accident description.
    """
    description = description.dropna()
    all_desc = []
    for desc in description:
        all_desc.append(desc)
        
    full_desc = " ".join(all_desc)
        
    main_words = WordCloud(scale=5,max_words=500).generate(full_desc)
    
    plt.figure(figsize=(22,14))
    
    
    plt.imshow(main_words,interpolation="bilinear")
    plt.axis("off")
    
    
    plt.title("Main Words in Accident Description",color='r')
    plt.show()
```

```python
clean_and_plot(df['Description'])
```

## <a id="conclusion"/> Conclusion

**Where do most accidents happen?**
    Most accidents happen in big states, as California, Texas, South Carolina, Florida, North Carolina and New York. The cities with most accidents also belong to these states, Houston, LA, Charlotte, Dallas and Austin. This could be due to large traffic, nonetheless we don't have those numbers in this dataset. The only solid conclusion we can make is that drivers should be extra careful in these localities.
    
**What are the surroundings of the accidents?**
    From our dataset, we saw that most accidents happen near traffic signals, juctions and crossing. The substancial difference to other traffic objects may lead us to conclude that these accidents happen due to a disrespect of traffic signals or the road code (in juctions). The high number of accidents near crossing may be related to excessive speed and sudden breaking, due to pedestrians. Again, we do not have enough data to confirm these assumptions, nevertheless drivers must be cautious near these road signs.
    
**When do most accidents happen?**
    As expected, most accidents happen at the hours of very high traffic. When people are going to work (6h to 11h) and when people are coming home (15h to 18h). Prudence is required when driving in these time frames.
    In terms of monthly data, the month with less accidents is July. This month corresponds to vacation time, which might explain why the numbers are lower.
    
**What relates to accident severity**
    Most people may anticipate that the weather conditions correlate with accident severity, however the numbers show it does not. No variable has a high correlation to accident severity. The variables that can be more connected to severity are the TMC, the Traffic Message Channel Code. A different code may be used when the severity of the accident is higher. The other variable is the distance of blocked road, which is intuitive, since more severe accidents may tend to require a higher extent of road blocking.
